```python
import sys
import vk_api
```

Authorization

```python
login, password = "username", "password"
vk = vk_api.VkApi(login, password)
    
try:
    vk.authorization()
except vk_api.AuthorizationError as error_msg:
    print(error_msg)
```

Get users ids from all user groups

```python
user_id = int(input())
    
users_groups = {}
groups = vk.method('groups.get', {'user_id': user_id,
                                     'extended': 1,
                                     'fields': [
                                                'name',
                                                'members_count',
                                                ]
                                    })['items']
groups_id = {}
for group in groups:
    groups_id[group['id']] = {
                               'name': group['name'],
                               'members_count': group['members_count']
                             }

for group_id in groups_id:
    members = vk.method('groups.getMembers', {'group_id': group_id})['items']
    for member in members:
        if member not in users_groups:
            users_groups[member] = {group_id}
        else:
            users_groups[member].add(group_id)
```

Get rank of each user and sorting

```python
user_rank = []
for user, group in users_groups.items():
    if user == user_id:
        continue
    mutual_groups = group.intersection(set(groups_id))
    user_rank += [(user, len(mutual_groups), map(lambda gid: groups_id[gid]['name'], mutual_groups))]
    
user_rank.sort(key=lambda x: -x[1])
```

```python
for i in range(10):
    print(user_rank[i][0])
    for gname in user_rank[i][2]:
        print(gname)
    print('\n')
```
